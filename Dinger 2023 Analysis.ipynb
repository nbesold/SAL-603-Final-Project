```python
# load in 2023 hitting stats
hitting2023 = pd.read_csv("2023 Hitting Stats.csv")
#create 2023 points column
hitting2023['UD Points'] = hitting2023['1B'] * 3 + hitting2023['2B'] * 6 + hitting2023['R'] * 2 + hitting2023['HR'] * 10 + hitting2023['RBI'] * 2 + hitting2023['SB'] * 4 + hitting2023['3B'] * 8 + hitting2023['BB'] * 3 + hitting2023['HBP'] * 3 + hitting2023['IBB'] * 3
#rename to match with draft data
hitting2023.rename(columns={'NameASCII': 'player_name'}, inplace=True)
#merge with reg season data
dinger_reg_season_stats = pd.merge(dinger_reg_season, hitting2023, on = ['player_name'], how = 'inner')
dinger_reg_season_stats
#merge with playoffs
dinger_reg_season_stats = pd.merge(dinger_reg_season, hitting2023, on = ['player_name'], how = 'inner')
dinger_first_round_stats = pd.merge(dinger_first_round, hitting2023, on = ['player_name'], how = 'inner')
dinger_second_round_stats = pd.merge(dinger_second_round, hitting2023, on = ['player_name'], how = 'inner')
dinger_finals_stats = pd.merge(dinger_finals, hitting2023, on = ['player_name'], how = 'inner')
```

```python
# Group draft picks by both "team_entry_id" and "team"
grouped_picks = dinger_reg_season_stats.groupby(['draft_entry_id', 'Team'])

# Count the number of players drafted by each fantasy team from each MLB team
team_player_counts = grouped_picks.size().reset_index(name='player_count')

# Count the occurrences of each number of players from the same MLB team
player_count_distribution = team_player_counts['player_count'].value_counts().sort_index()

# Calculate the total number of fantasy teams
total_teams = dinger_reg_season_stats['draft_entry_id'].nunique()

# Calculate the percentage for each number of players from the same MLB team
percentages = (player_count_distribution / total_teams) * 100

# Print the results
for count, percentage in percentages.items():
    print(f"Percentage of fantasy teams with {count} players from the same MLB team: {percentage:.2f}%")
```

```python
# look into stacks playoffs
grouped_picks_1st = dinger_first_round_stats.groupby(['draft_entry_id', 'Team'])
team_player_counts_1st = grouped_picks_1st.size().reset_index(name='player_count')
# Filter teams with 4 or more players
teams_with_4_or_more_players_1st = team_player_counts_1st[team_player_counts_1st['player_count'] >= 4]

# Count the number of fantasy teams with 4 or more players from the same MLB team
num_teams_with_4_or_more_players_1st = len(teams_with_4_or_more_players_1st['draft_entry_id'].unique())

# Total number of fantasy teams
total_teams_1st = dinger_first_round_stats['draft_entry_id'].nunique()

# Calculate the percentage
percentage_1st = (num_teams_with_4_or_more_players_1st / total_teams_1st) * 100
percentage_1st
```

```python
# look into stacks playoffs
grouped_picks_finals = dinger_finals_stats.groupby(['draft_entry_id', 'Team'])
team_player_counts_finals = grouped_picks_finals.size().reset_index(name='player_count')
# Filter teams with 4 or more players
teams_with_4_or_more_players_finals = team_player_counts_finals[team_player_counts_finals['player_count'] >= 4]

# Count the number of fantasy teams with 4 or more players from the same MLB team
num_teams_with_4_or_more_players_finals = len(teams_with_4_or_more_players_finals['draft_entry_id'].unique())

# Total number of fantasy teams
total_teams_finals = dinger_finals_stats['draft_entry_id'].nunique()

# Calculate the percentage
percentage_finals = (num_teams_with_4_or_more_players_finals / total_teams_finals) * 100
percentage_finals
```
