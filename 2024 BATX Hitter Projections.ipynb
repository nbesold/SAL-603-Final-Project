```python
# import pandas and numpy
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import unicodedata

```

```python
# load in batx projections
projections = pd.read_csv("2024 BATX Hitter Projections.csv")
pitcher_projections = pd.read_csv("2024 BAT Pitcher Projections.csv")
underdog_adp = pd.read_csv("Underdog ADP Data.csv")
underdog_adp
```

```python
# Create points column for both pitchers/hitters
projections['UD Points'] = projections['1B'] * 3 + projections['2B'] * 6 + projections['R'] * 2 + projections['HR'] * 10 + projections['RBI'] * 2 + projections['SB'] * 4 + projections['3B'] * 8 + projections['BB'] * 3 + projections['HBP'] * 3 + projections['IBB'] * 3
pitcher_projections['UD Pitcher Points'] = pitcher_projections['W'] * 5 + pitcher_projections['SO'] * 3 + pitcher_projections['QS'] * 5 + pitcher_projections['ER'] * -(3) + pitcher_projections['IP'] * 3
projections
```

```python
# merge underdog firstname lastname columns, drop afterwards 
underdog_adp['Name'] = underdog_adp['firstName'] + ' ' + underdog_adp['lastName']
underdog_adp.drop(columns=['firstName', 'lastName'], inplace=True)

underdog_adp
```

```python
# function to normalize accents in projections      Might not need anympre
def normalize_accents(text):
    return ''.join(char for char in unicodedata.normalize('NFD', text)
                   if unicodedata.category(char) != 'Mn')

# Normalize accents in the "name" column of both DataFrames
projections['Name1'] = projections['Name'].apply(normalize_accents)
pitcher_projections['Name1'] = pitcher_projections['Name'].apply(normalize_accents)
underdog_adp['Name1'] = underdog_adp['Name'].apply(normalize_accents)

#drop Jr. from name column
projections['Name1'] = projections['Name1'].str.replace(' Jr.', '')
pitcher_projections['Name1'] = pitcher_projections['Name1'].str.replace(' Jr.', '')
```

```python
# merge hitting data frame to adp
ud_projections = projections.merge(underdog_adp[['Name1', 'positionRank', 'adp']], on='Name1', how='left')
ud_projections
ud_projections.to_csv("UD Batter Projections.csv")
```

```python
# merge pitching dataframe to ud adp
ud_pitcher_projections = pitcher_projections.merge(underdog_adp[['Name1', 'positionRank', 'adp']], on='Name1', how='left')
ud_pitcher_projections
ud_pitcher_projections.to_csv("UD Pitcher Projections.csv")
```

```python
# fix decimal places in pitchers
ud_pitcher_projections = ud_pitcher_projections.round(2)
ud_pitcher_projections
ud_pitcher_projections.to_csv("UD Pitcher Projections.csv")
```
